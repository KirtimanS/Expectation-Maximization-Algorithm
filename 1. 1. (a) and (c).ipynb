```python
import math
import numpy as np
def gaussian(x,mean,var):
    return math.exp(-(((x-mean)/var)**2)/2)*(1/((2*math.pi*(var**2))**(0.5)))
def model(x,m1,m2):
    log_likelihood = 0
    for point in x:
        log_likelihood += math.log(0.9*(0.7*gaussian(point,m1,2) + 0.3*gaussian(point,m2,2)) + 0.0025)
#         print(likelihood)
    return log_likelihood
   
```

```python
coord = []
for mean1 in range(-20,21):
    for mean2 in range(-20,21):
        coord.append([model(data, mean1, mean2), mean1, mean2])
```
